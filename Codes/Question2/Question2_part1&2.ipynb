```python
import numpy as np
import matplotlib.pyplot as plt
import random

size = 100
S1_x1 = np.random.normal(loc = 0, scale = 0.3, size = size)
S1_y1 = np.random.normal(loc = 0, scale = 0.3, size = size)

S1_x2 = np.random.normal(loc = 2, scale = 0.3, size = size)
S1_y2 = np.random.normal(loc = 2, scale = 0.3, size = size)

plt.plot(S1_x1, S1_y1, 'ro', label = 'class1')
plt.plot(S1_x2, S1_y2, 'bo', label = 'class2')
plt.legend(loc = "upper right")
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

w1 = random.random()
w2 = random.random()
b = random.random()
alpha = 0.1
weight = [w1, w2]

t = [1, -1]
X = [S1_x1, S1_x2]
Y = [S1_y1, S1_y2]
thershold = 0.000001

cost = []
point = []

for j in range(2):
    for i in range(size):
        point.append([X[j][i], Y[j][i], t[j]])

error = 1

while error > thershold:
    c = random.choice(point)
    net = b + np.dot(c[0:2], weight)
    w1 = w1 + alpha * (c[-1] - net) * c[0]
    w2 = w2 + alpha * (c[-1] - net) * c[1]
    weight = [w1, w2]
    b = b + alpha * (c[-1] - net)
    error = 0.5 * (c[-1] - net) ** 2
    cost.append(error)



x_line = np.arange(-1, 3, 0.5)
line = -(b)/(weight[1]) - ((weight[0])/(weight[1]))*x_line

plt.plot(S1_x1, S1_y1, 'ro', label = 'class1')
plt.plot(S1_x2, S1_y2, 'bo', label = 'class2')
plt.legend(loc = "upper right")

plt.plot(x_line, line)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Classifying')
plt.show()


plt.plot(cost)
plt.xlabel('Number of tries')
plt.ylabel('error')
plt.show()

type(line)
```
