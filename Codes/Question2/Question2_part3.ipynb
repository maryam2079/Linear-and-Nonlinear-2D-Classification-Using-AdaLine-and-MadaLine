```python
import numpy as np
import matplotlib.pyplot as plt
import random

size1 = 100
size2 = 20

x1 = np.random.normal(loc = 2, scale = 1, size = size1)
y1 = np.random.normal(loc = 1, scale = 0.1, size = size1)

x2 = np.random.normal(loc = -1, scale = 0.4, size = size2)
y2 = np.random.normal(loc = 2, scale = 0.4, size = size2)

w1 = random.random()
w2 = random.random()
b = random.random()
alpha = 0.1
weight = [w1, w2]

t = [1, -1]

thershold = 0.000001

cost = []
point = []

for i in range(size1):
    point.append([x1[i], y1[i], t[0]])
for i in range(size2):
    point.append([x2[i], y2[i], t[1]])
    
error = 1

while error > thershold:
    c = random.choice(point)
    net = b + np.dot(c[0:2], weight)
    w1 = w1 + alpha * (c[-1] - net) * c[0]
    w2 = w2 + alpha * (c[-1] - net) * c[1]
    weight = [w1, w2]
    b = b + alpha * (c[-1] - net)
    error = 0.5 * (c[-1] - net) ** 2
    cost.append(error)



x_line = np.arange(-1, 5, 0.5)
line = -(b)/(weight[1]) - ((weight[0])/(weight[1]))*x_line

plt.plot(x1, y1, 'ro', label = 'class1')
plt.plot(x2, y2, 'bo', label = 'class2')
plt.legend(loc = "upper right")

plt.plot(x_line, line)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Classifying')
plt.show()


plt.plot(cost)
plt.xlabel('Number of tries')
plt.ylabel('error')
plt.show()
```
